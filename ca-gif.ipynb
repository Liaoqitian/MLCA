```python
def count_neighbors(data, i, j, type="Moore"):
    max_state = number_states - 1
    if type == "Moore": 
        res = 0
        if i > 0 and data[i - 1][j] == max_state:
            res += 1
        if i < len(data) - 1 and data[i + 1][j] == max_state:
            res += 1
        if j > 0 and data[i][j - 1] == max_state:
            res += 1
        if j < len(data[0]) - 1 and data[i][j + 1] == max_state:
            res += 1
        if i > 0 and j > 0 and data[i - 1][j - 1] == max_state: 
            res += 1
        if i > 0 and j < len(data[0]) - 1 and data[i - 1][j + 1] == max_state:
            res += 1
        if i < len(data) - 1 and j > 0 and data[i + 1][j - 1] == max_state:
            res += 1
        if i < len(data) - 1 and j < len(data[0]) - 1 and data[i + 1][j + 1] == max_state:
            res += 1
        return res 
    elif type == "Neumann":
        res = 0
        if i > 0 and data[i - 1][j] == max_state:
            res += 1
        if i < len(data) - 1 and data[i + 1][j] == max_state:
            res += 1
        if j > 0 and data[i][j - 1] == max_state:
            res += 1
        if j < len(data[0]) - 1 and data[i][j + 1] == max_state:
            res += 1
        return res 
    else: 
        raise Exception("Invalid Argument")
    return res
```

```python
def evolve(data, survival_arr, born_arr): 
    copy = [[0 for j in range(length)] for i in range(width)]
    max_state = number_states - 1
    for i in range(width):
        for j in range(length):
            if data[i][j] > 0 and data[i][j] < max_state: ## sick cells get sicker
                copy[i][j] = data[i][j] - 1
            else: 
                count = count_neighbors(data, i, j)
                if data[i][j] == max_state and count in survival_arr: ## live cells stay alive
                    copy[i][j] = data[i][j]
                if data[i][j] == max_state and count not in survival_arr: ## live cells become sick
                    copy[i][j] = data[i][j] - 1
                elif data[i][j] == 0 and count in born_arr: ## dead cells become alive
                    copy[i][j] = max_state
    return copy
```
